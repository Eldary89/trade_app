```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
```

```python
data=pd.read_csv('eurgbp_d.csv')
```

```python
print(data.head(10))
```

```python
#high-low
data['h-l']= data['high'] - data['low']

data.head()
#entry level buy orders
data['lev1b']= data['high']- data['h-l']*0.236
data['lev2b']= data['high']- data['h-l']*0.382
data['lev3b']= data['high']- data['h-l']*0.50
data['lev4b']= data['high']- data['h-l']*0.618
data['lev5b']= data['high']- data['h-l']*0.7647

#entry level sell orders
data['lev1s']= data['low']+ data['h-l']*0.236
data['lev2s']= data['low']+ data['h-l']*0.382
data['lev3s']= data['low']+ data['h-l']*0.50
data['lev4s']= data['low']+ data['h-l']*0.618
data['lev5s']= data['low']+ data['h-l']*0.7647

```

```python
data['decision'] = np.where(data['h-l'] > data['atr'], 'trade', 'not')
print(data.head(10))
```

```python
data['trade'] = np.where(data['decision'] == 'trade', np.where(data['open'] < data['close'], 'buy', 'sell'), 'nothing')
print(data.head(10))
```

```python
data['Buy'] = np.where(data['trade'] == 'buy', 
                       data['lev1b'] * 0.2 + data['lev2b'] * 0.2 + data['lev3b'] * 0.2 + data['lev4b'] * 0.2 + data['lev5b'] * 0.2,
                       0)
print(data.head(10))
```

```python
data['Sell'] = np.where(data['trade'] == 'sell', 
                       data['lev1s'] * 0.2 + data['lev2s'] * 0.2 + data['lev3s'] * 0.2 + data['lev4s'] * 0.2 + data['lev5s'] * 0.2,
                       0)
print(data.head(10))
```
