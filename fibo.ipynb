```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
```

```python
data=pd.read_csv('eurgbp_d.csv')
```

```python
print(data.head(10))
```

```python
#high-low
data['h-l']= data['high'] - data['low']

data.head()
#entry level buy orders
data['lev1b']= data['high']- data['h-l']*0.236
data['lev2b']= data['high']- data['h-l']*0.382
data['lev3b']= data['high']- data['h-l']*0.50
data['lev4b']= data['high']- data['h-l']*0.618
data['lev5b']= data['high']- data['h-l']*0.7647

#entry level sell orders
data['lev1s']= data['low']+ data['h-l']*0.236
data['lev2s']= data['low']+ data['h-l']*0.382
data['lev3s']= data['low']+ data['h-l']*0.50
data['lev4s']= data['low']+ data['h-l']*0.618
data['lev5s']= data['low']+ data['h-l']*0.7647





```

```python
data['decision'] = np.where(data['h-l'] > data['atr'], 'trade', 'not')
print(data.head(10))
```

```python
data['trade'] = np.where(data['decision'] == 'trade', np.where(data['open'] < data['close'], 'buy', 'sell'), 'nothing')
print(data.head(10))
```

```python
data['Buy_Lev1'] = np.where(data['trade'].shift(1) == 'buy', 
                            np.where(data['low'] <= data['lev1b'].shift(1), 
                                     np.where(data['lev1b'].shift(1) <= data['high'], 
                                              data['close'] - data['lev1b'].shift(1), 0), 0), 0)

data['Buy_Lev2'] = np.where(data['trade'].shift(1) == 'buy', 
                            np.where(data['low'] <= data['lev2b'].shift(1), 
                                     np.where(data['lev2b'].shift(1) <= data['high'], 
                                              data['close'] - data['lev2b'].shift(1), 0), 0), 0)

data['Buy_Lev3'] = np.where(data['trade'].shift(1) == 'buy', 
                            np.where(data['low'] <= data['lev3b'].shift(1), 
                                     np.where(data['lev3b'].shift(1) <= data['high'], 
                                              data['close'] - data['lev3b'].shift(1), 0), 0), 0)

data['Buy_Lev4'] = np.where(data['trade'].shift(1) == 'buy', 
                            np.where(data['low'] <= data['lev4b'].shift(1), 
                                     np.where(data['lev4b'].shift(1) <= data['high'], 
                                              data['close'] - data['lev4b'].shift(1), 0), 0), 0)

data['Buy_Lev5'] = np.where(data['trade'].shift(1) == 'buy', 
                            np.where(data['low'] <= data['lev5b'].shift(1), 
                                     np.where(data['lev5b'].shift(1) <= data['high'], 
                                              data['close'] - data['lev5b'].shift(1), 0), 0), 0)
print(data.head(10))
```

```python
data['Sell_Lev1'] = np.where(data['trade'].shift(1) == 'sell', 
                            np.where(data['low'] <= data['lev1s'].shift(1), 
                                     np.where(data['lev1s'].shift(1) <= data['high'], 
                                              data['close'] - data['lev1s'].shift(1), 0), 0), 0)

data['Sell_Lev2'] = np.where(data['trade'].shift(1) == 'sell', 
                            np.where(data['low'] <= data['lev2s'].shift(1), 
                                     np.where(data['lev2s'].shift(1) <= data['high'], 
                                              data['close'] - data['lev2s'].shift(1), 0), 0), 0)

data['Sell_Lev3'] = np.where(data['trade'].shift(1) == 'sell', 
                            np.where(data['low'] <= data['lev3s'].shift(1), 
                                     np.where(data['lev3s'].shift(1) <= data['high'], 
                                              data['close'] - data['lev3s'].shift(1), 0), 0), 0)

data['Sell_Lev4'] = np.where(data['trade'].shift(1) == 'sell', 
                            np.where(data['low'] <= data['lev4s'].shift(1), 
                                     np.where(data['lev4s'].shift(1) <= data['high'], 
                                              data['close'] - data['lev4s'].shift(1), 0), 0), 0)

data['Sell_Lev5'] = np.where(data['trade'].shift(1) == 'sell', 
                            np.where(data['low'] <= data['lev5s'].shift(1), 
                                     np.where(data['lev5s'].shift(1) <= data['high'], 
                                              data['close'] - data['lev5s'].shift(1), 0), 0), 0)
print(data.head(10))
```

```python

data['lev1btp'] = data['high'] + data['h-l'] *1.236
data['lev2btp'] = data['high'] + data['h-l']*1.382
data['lev3btp'] = data['high'] + data['h-l']*1.5
data['lev4btp'] = data['high'] + data['h-l']*1.618
data['lev5btp'] = data['high'] + data['h-l']*1.7647
print(data.head(10))
```

```python
# Check if range inside levbtp then subtract levbtp - levb
data['Dif_lev1btp_lev1b'] = np.where(data['low'] <= data['lev1btp'],
                                     np.where(data['lev1btp'] <= data['high'], data['lev1btp'] - data['lev1b'], 0), 0)
data['Dif_lev2btp_lev2b'] = np.where(data['low'] <= data['lev2btp'],
                                     np.where(data['lev2btp'] <= data['high'], data['lev2btp'] - data['lev2b'], 0), 0)
data['Dif_lev3btp_lev3b'] = np.where(data['low'] <= data['lev3btp'],
                                     np.where(data['lev3btp'] <= data['high'], data['lev3btp'] - data['lev3b'], 0), 0)
data['Dif_lev4btp_lev4b'] = np.where(data['low'] <= data['lev4btp'],
                                     np.where(data['lev4btp'] <= data['high'], data['lev4btp'] - data['lev4b'], 0), 0)
data['Dif_lev5btp_lev5b'] = np.where(data['low'] <= data['lev5btp'],
                                     np.where(data['lev5btp'] <= data['high'], data['lev5btp'] - data['lev5b'], 0), 0)
print(data.head(10))
```

```python
data['buy_loss_lev1'] = data['low'] - data['lev1b']
data['buy_loss_lev2'] = data['low'] - data['lev2b']
data['buy_loss_lev3'] = data['low'] - data['lev3b']
data['buy_loss_lev4'] = data['low'] - data['lev4b']
data['buy_loss_lev5'] = data['low'] - data['lev5b']
print(data.head(10))
```

```python
data['lev1stp']= data['lev1s']- data['h-l']
data['lev2stp']= data['lev2s']- data['h-l']
data['lev3stp']= data['lev3s']- data['h-l']
data['lev4stp']= data['lev4s']- data['h-l']
data['lev5stp']= data['lev5s']- data['h-l']
print(data.head(10))
```

```python
# The same for sell but levs - levstp
data['Dif_lev1stp_lev1s'] = np.where(data['low'] <= data['lev1stp'],
                                     np.where(data['lev1stp'] <= data['high'], data['lev1s'] - data['lev1stp'], 0), 0)
data['Dif_lev2stp_lev2s'] = np.where(data['low'] <= data['lev2stp'],
                                     np.where(data['lev2stp'] <= data['high'], data['lev2s'] - data['lev2stp'], 0), 0)
data['Dif_lev3stp_lev3s'] = np.where(data['low'] <= data['lev3stp'],
                                     np.where(data['lev3stp'] <= data['high'], data['lev3s'] - data['lev3stp'], 0), 0)
data['Dif_lev4stp_lev4s'] = np.where(data['low'] <= data['lev4stp'],
                                     np.where(data['lev4stp'] <= data['high'], data['lev4s'] - data['lev4stp'], 0), 0)
data['Dif_lev5stp_lev5s'] = np.where(data['low'] <= data['lev5stp'],
                                     np.where(data['lev5stp'] <= data['high'], data['lev5s'] - data['lev5stp'], 0), 0)
print(data.head(10))
```

```python
data['sell_loss_lev1'] = data ['lev1b'] - data ['high']
data['sell_loss_lev2'] = data ['lev2b'] - data ['high']
data['sell_loss_lev3'] = data ['lev3b'] - data ['high']
data['sell_loss_lev4'] = data ['lev4b'] - data ['high']
data['sell_loss_lev5'] = data ['lev5b'] - data ['high']
print(data.head(10))
```

```python
data['pnl_lev1b']=data['lev1btp']-data['lev1b']
data['pnl_lev2b']=data['lev2btp']-data['lev2b']
data['pnl_lev3b']=data['lev3btp']-data['lev3b']
data['pnl_lev4b']=data['lev4btp']-data['lev4b']
data['pnl_lev5b']=data['lev5btp']-data['lev5b']
print(data.head(10))
```

```python
data['pnl_lev1s']=data['lev5s'] -data['lev1stp']
data['pnl_lev2s']=data['lev4s'] -data['lev2stp']
data['pnl_lev3s']=data['lev3s'] -data['lev3stp']
data['pnl_lev4s']=data['lev2s'] -data['lev4stp']
data['pnl_lev5s']=data['lev1s'] -data['lev5stp']
print(data.head(10))
```
